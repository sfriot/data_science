## <font color=blue>1. Approche utilisée</font>

Pour ce projet, j'ai commencé par me documenter pour accroître ma connaissance métier. Je me suis concentré particulièrement sur les critères pour choisir un pays lorsqu'une société veut développer une activité d'exportation.  
Il faut garder en tête que notre société souhaite dans un premier temps seulement exporter, et non pas développer une activité de production sur place.  
En croisant mes différentes recherches, je dresse un tableau des critères à potentiellement prendre en compte. Il est important de noter que nous ne sommes pas obligés d'analyser tous les critères de ce tableau, car ce tableau est assez complet et certains critères se croisent.

<a id="1_1"></a>

### <font color=blue>1.1. Critères de l'étude de marché</font>

**Tableau des critères de sélection d'un pays pour exporter**  
  
**Domaine** | **Indicateur** | **Appréciation**  
:-: | :-: | :-:  
Environnement géographique | Distance par rapport au marché local | Plus la distance est grande, plus les frais de logistique et de transport risquent d'être élevés.  
 | Conditions climatiques | Certains climats nécessitent un emballage ou un stockage spécifiques.  
 |  
Démographie | Population totale et sa croissance | Détermine le potentiel d'un marché et son évolution.  
 | Densité et concentration de la population | Influence les coûts de distribution et de communication.  
 | Degré d'urbanisation | Les populations urbaines et rurales présentent des différences culturelles et économiques importantes. C'est un indicateur indirect de la densité et concentration de la population.  
 |  
Infrastructures | Infrastructures de transport | Leur nombre et leur qualité assurent un transport de qualité et à un coût modéré.  
 | Etat des circuits de distribution et de stockage | Important dans le cas de commercialisation de produits fragiles et périssables.  
 | Moyens de communication | Indicateur de dévelopemment du pays et de possibilité de travailler dans de bonnes conditions avec des partenaires locaux.  
 |  
Economie | PIB globale et par habitant | Renseigne sur le niveau de vie et le pouvoir d'achat des habitants.  
 | Distribution des revenus | Elle permet de savoir si la moyenne du PIB par habitant cache de grandes disparités, et donc des opportunités.  
 | Situation et sécurité économique | La situation économique globale, et la sécurité des affaires, sont des points qui influent sur l'opportunité de développer des exportations vers un pays donné.  
 | Données du secteur d'activité dans le pays ciblé | L'existence préalable d'une consommation dans le pays permet de s'assurer que le marché existe déjà et qu'il est d'une taille suffisamment importante pour s'y intéresser.  
 | Données du commerce extérieur du secteur d'activité | La production, les importations, les exportations et le solde de la balance commerciale donnent des indications générales sur le degré d'ouverture du marché et l'état de l'industrie et de la demande locales.  
 | Appartenance à un accord régionale | Peut rendre plus facile ou difficile les exportations vers ce pays.  
 | Politique fiscale et les taxes | Une fiscalité élevée peut constituer une barrière à l'entrée, en augmentant fortement le prix final du bien vendu.  
 |  
Environnement politique | Stabilité politique | Un régime politique instable est défavorable au développement des activités économiques et commerciales.  
 | Organisation politique et administrative | Le niveau de bureaucratie et son organisation efficace influencent le climat des affaires et la facilité administrative à développer une activité.

Certains critères peuvent être évalués avec différents indicateurs, qui offrent des angles d'analyse variés.  
Parfois, le même indicateur est disponible dans différents fichiers. Dans cette hypothèse, je le télécharge aux différents endroits où il est disponible au cas où l'une des sources ait des valeurs manquantes et pour vérifier éventuellement la concordance des informations.

**Deuxième source de données : le site Doing Business de La Banque Mondiale**  
  
Le site http://francais.doingbusiness.org/fr/data/doing-business-score propose de télécharger une calculatrice Excel qui calcule un score de facilité pour développer une activité dans les différents pays du monde, à fin avril 2019. Ce score est basé sur :  
- le coût en argent et en temps des démarches administratives dans sept domaines : démarrer une activité, la gestion des permis de construire, l'alimentation en électricité, l'enregistrement des propriétés foncières, l'obtention d'un crédit, le paiement des taxes (montant des taxes et facilité des tâches administratives), le commerce international (montant des taxes et facilité des taches administratives à l'importation et à l'exportation) ;  
- la sécurité des investissements et des relations commerciales dans trois domaines : la protection des investisseurs minoritaires, les procédures judiciaires en cas de non-respect d'un contrat, et le recouvrement des impayés.  
  
Dans notre cas, nous souhaitons seulement exporter, pas implanter une activité de production sur place. Par conséquent, j'ai adapté la calculatrice Excel pour calculer un score basé uniquement sur le démarrage de l'activité, le paiement des taxes, le commerce international du seul point de vue des importations, le respect des contrats et le recouvrement des impayés. Puis j'ai ajouté une colonne avec le code pays de la FAO (grâce au code ISO3).  
  
L'ensemble de ces évaluations aboutit à une note entre 0 et 100 qui évalue la facilité et la sécurité de mettre en place et maintenir des exportations vers les différents pays. Le fichier projet05_opportunites_doingbusiness.csv contient cet indicateur synthétique de la sécurité économique et de l'organisation politique et administrative du pays.

**Troisième source de données : le site de Credendo, groupe d'assurance-crédit internationale**
  
Le site https://www.credendo.com/fr/country-risk permet de télécharger un fichier Excel qui contient l'estimation par Credendo des risques politiques et commerciaux pour les exportations vers les différents pays du monde. J'ai ajouté une colonne avec le code pays de la FAO (grâce au code ISO3). Je retiens deux données :  
- le risque politique des transactions, qui est une synthèse des risques politiques à court terme et à moyen-long terme. Cette note va de 1 à 7, la note la plus faible correspondant au risque le plus faible. C'est une donnée quantitative, mais elle peut être vue comme une donnée catégorielle.  
- le risque commercial des exportations : cette note va de 1 à 3 et évalue le risque économique d'exporter vers ce pays (incertitude quant au respect des contrats et au paiement). La note la plus faible correspond au risque le plus faible. C'est une donnée quantitative, mais elle peut être vue comme une donnée catégorielle.  
  
Ces deux notes sont de nouveaux indicateurs de la sécurité économique et politique du pays. Mais leur caractère catégoriel ne permet peut-être pas une utilisation optimale dans le cadre d'une ACP.

**Dernière donnée ajoutée : la distance géographique**
  
Le site https://www.annuaire-mairie.fr/distance-capitale-paris.html propose la distance à vol d'oiseau entre Paris et toutes les capitales du monde. J'ai scrappé cette page pour créer un fichier csv qui comporte le nom du pays, et la distance de capitale à capitale entre ce pays et la France.

```python
temp = pd.read_csv("projet05_FAOSTAT_population.csv")
temp.head()
```

```python
temp.columns = ["xx1","xx2","country_code","country","xx3","stat","xx4","xx5","xx6",\
                      "year","unit","value","xx7","xx8","xx9"]
temp.drop(columns=["xx1","xx2","xx3","xx4","xx5","xx6","xx7","xx8","xx9"], inplace=True)
temp.describe(include='all')
```

L'unité est de 1000 personnes pour toutes les lignes. Je multiplie value par 1000 et supprime la colonne unit.  
Je crée un tableau croisé dynamique avec les pays, les codes pays et l'année en index et la stat en colonnes.

```python
temp.value = temp.value * 1000
temp.drop(columns="unit", inplace=True)
population = temp.pivot_table(index=["country_code","country","year"], columns="stat", values="value", aggfunc=sum)
population.reset_index(inplace=True)
population.head()
```

```python
population.columns = ["country_code","country","year","rural_population","total_population","urban_population"]
population.isna().sum()
```

J'ajoute une colonnes dans laquelle je calcule l'évolution de la population. Je choisis de faire le calcul sur 10 ans. Pour cela, je crée un dataframe temporaire où je décale les données de 10 ans.

```python
nb_year = 10
population.sort_values(by=["country_code","year"], inplace=True)
temp = population.shift(periods=nb_year)
for i, row in population.iterrows():
    if population.loc[i, "country_code"] == temp.loc[i, "country_code"]:
        population.loc[i, "croissance_population_10yr"] = (population.loc[i, "total_population"] / temp.loc[i, "total_population"]) - 1.0
    else:
        population.loc[i, "croissance_population_10yr"] = np.nan
```

La Serbie-Monténégro a été divisée en 2 pays, la Serbie et et le Monténégro, en juin 2006 à la suite du référendum sur l'indépendance du Monténégro du 21 mai 2006 (source Wikipedia). Pour les années 2005 et précédentes, je dispose des données de la Serbie-Monténégro. Je vais calculer le ratio de la population entre la Serbie et le total des deux pays en 2006 pour en déduire une estimation de la population de la Serbie et du Monténégro pour les années précédentes. Je peux ainsi calculer une estimation de l'évolution de la croissance de la population sur 10 ans.  
  
Je fais de même pour le Soudan (ex) qui a été séparé en 2 entités, le Soudan et le Soudan du Sud, en juillet 2011 à la suite du référendum sur l'indépendance du Soudan du Sud du 9 au 15 janvier 2011 (source Wikipedia). Pour les années 2011 et précédentes, je dispose des données pour le Soudan (ex) qui me permettent d'estimer la croissance pour chacun des deux pays à partir de 2012.

```python
serbie_population2006 = population[(population.country_code==272) & (population.year==2006)].total_population.values[0]
montenegro_population2006 = population[(population.country_code==273) & (population.year==2006)].total_population.values[0]
serbie_ratio2006 = serbie_population2006 / (serbie_population2006 + montenegro_population2006)
interpopulation = []
for annee in np.arange(1996,2006):
    temp1 = np.trunc(population[(population.country=="Serbie-et-Monténégro") & (population.year==annee)].total_population.values[0] * serbie_ratio2006)
    interpopulation.append([272,"Serbie",annee,temp1])
    temp2 = population[(population.country=="Serbie-et-Monténégro") & (population.year==annee)].total_population.values[0] - temp1
    interpopulation.append([273,"Monténégro",annee,temp2])
for annee in np.arange(2006,2018):
    if np.isnan(population[(population.country=="Serbie") & (population.year==annee)].croissance_population_10yr.values[0]):
        temp1 = population[(population.country=="Serbie") & (population.year==annee)].total_population.values[0]
        interpopulation.append([272,"Serbie",annee,temp1])
        temp2 = population[(population.country=="Monténégro") & (population.year==annee)].total_population.values[0]
        interpopulation.append([273,"Monténégro",annee,temp2])
```

```python
soudan_population2012 = population[(population.country_code==276) & (population.year==2012)].total_population.values[0]
sud_population2012 = population[(population.country_code==277) & (population.year==2012)].total_population.values[0]
soudan_ratio2012 = soudan_population2012 / (soudan_population2012 + sud_population2012)
for annee in np.arange(2002,2012):
    temp1 = np.trunc(population[(population.country=="Soudan (ex)") & (population.year==annee)].total_population.values[0] * soudan_ratio2012)
    interpopulation.append([276,"Soudan",annee,temp1])
    temp2 = population[(population.country=="Soudan (ex)") & (population.year==annee)].total_population.values[0] - temp1
    interpopulation.append([277,"Soudan du Sud",annee,temp2])
for annee in np.arange(2012,2018):
    if np.isnan(population[(population.country=="Soudan") & (population.year==annee)].croissance_population_10yr.values[0]):
        temp1 = population[(population.country=="Soudan") & (population.year==annee)].total_population.values[0]
        interpopulation.append([276,"Soudan",annee,temp1])
        temp2 = population[(population.country=="Soudan du Sud") & (population.year==annee)].total_population.values[0]
        interpopulation.append([277,"Soudan du Sud",annee,temp2])
interpopulation = pd.DataFrame(interpopulation, columns=["country_code","country","year","total_population"])
```

Je décide de conserver les données en quantité plutôt qu'en valeur lorsque j'ai le choix entre ces deux approches. Pour les données en dollars, je préfère garder les données en dollar courant plutôt que les données en dollar constant car ces dernières sont plutôt utilisées pour comparer les évaluations dans le temps.  
  
Pour mon premier dataframe, je garde les données suivantes :  
- la croissance de la population ;  
- la proportion de protéines d'origine animale dans la disponibilité alimentaire = animals_protein_supply_g_capita_day / (animals_protein_supply_g_capita_day + vegetals_protein_supply_g_capita_day) ;  
- la disponibilité alimentaire en protéines par habitant = animals_protein_supply_g_capita_day + vegetals_protein_supply_g_capita_day ;  
- la disponibilité alimentaire en kilo-calories par habitant = animals_food_supply_kcal_capita_day + vegetals_food_supply_kcal_capita_day ;  
- la proportion de protéines provenant de la viande de volaille dans les protéines d'origine animale = poultry_protein_supply_g_capita_day / animals_protein_supply_g_capita_day ;  
- la balance commerciale de la viande de volaille = poultry_export_tonnes - poultry_import_tonnes ;  
- la disponibilité alimentaire en viande de volaille = poultry_food_tonnes ;  
- la proportion des importations dans la disponibilité alimentaire de viande de volaille = poultry_import_tonnes / poultry_food_tonnes ;  
- le taux d'urbanisation = ratio urban_population / total_population ;  
- l'indice de stabilité politique de la FAO = political stability ;  
- le PIB par habitant (je sélectionne le PIB par habitant qui est plus représentatif du pouvoir d'achat individuel que le PIB du pays) : gdp_per_capita_usd ;  
- la distribution des revenus (répartition égalitaire ou concentrée) = food_supply_variability_kcal_capita_day / (animals_food_supply_kcal_capita_day + vegetals_food_supply_kcal_capita_day) ;  
- le score Doing Business qui évalue les risques et opportunités des exportations vers le pays : db_export_score.  
  
Pour mon deuxième dataframe, je stocke les informations complémentaires suivantes :  
- la population : total_population ;  
- la distance de Paris de la capitale du pays : distance_capitale_paris ;  
- le score de risque politique de Credendo, doublon de l'indice de stabilité politique de la FAO : credendo_political_risk ;  
- le score de risque commercial de Credendo, doublon du score Doing Business : credendo_commercial_risk ;  
- le PIB : gdp_usd ;  
- la proportion des importations par rapport à la production de viande de volaille = poultry_import_tonnes / poultry_production_tonnes ;  
- les données en quantité des importations et de la production pour le poulet, le poulet en boîte et la viande de volaille : chicken_imports_quantity_number, poultry_imports_quantity_tonnes, cannedchicken_imports_quantity_tonnes, chicken_stock_number, poultry_production_quantity_tonnes, chicken_production_quantity_tonnes ;
